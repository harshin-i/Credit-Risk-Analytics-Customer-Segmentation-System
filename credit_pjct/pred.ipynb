```python
import pandas as pd
import pickle
```

```python
file_path = r"C:\Users\Sriharshini\Downloads\nbfc_credit_risk_dataset.csv"
df = pd.read_csv(file_path)
```

```python
with open("models/risk_model.pkl", "rb") as f:
    model = pickle.load(f)

with open("models/scaler.pkl", "rb") as f:
    scaler = pickle.load(f)
```

```python
model_features = [
    "age",
    "years_in_operation",
    "annual_revenue",
    "ebitda_margin",
    "cashflow_stability",
    "bank_balance_avg",
    "dti_ratio",
    "dscr",
    "loan_amount",
    "tenure_months",
    "interest_rate",
    "past_delinquencies",
    "dpd_last_6m",
    "repayment_consistency",
    "monthly_inflow",
    "monthly_outflow",
    "default"       
]
```

```python
df["default"] = df["default"].fillna(0)
```

```python
X = df[model_features]
X_scaled = scaler.transform(X)
```

```python
X_scaled_final = X_scaled.copy()
```

```python
y_pred = model.predict(X_scaled)
```

```python
print(scaler.feature_names_in_)
```

```python
df["predicted_probability"] = model.predict_proba(X_scaled)[:, 1]
```

```python
def assign_risk_tier(prob):
    if prob < 0.3:
        return "Low"
    elif prob < 0.6:
        return "Medium"
    else:
        return "High"

df["risk_tier"] = df["predicted_probability"].apply(assign_risk_tier)
```

```python
seg = pd.read_csv("segmentation_output.csv")
df = df.merge(seg[["borrower_id", "cluster_id"]], on="borrower_id", how="left")
```

```python
pred = df[["borrower_id", "predicted_probability", "risk_tier", "default"]]
pred.to_csv("predictions.csv", index=False)

print("predictions.csv saved!")
```
