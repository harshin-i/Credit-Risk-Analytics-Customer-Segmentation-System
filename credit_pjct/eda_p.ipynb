```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set(style="whitegrid")
plt.rcParams["figure.figsize"] = (10, 6)
```

```python
file_path = r"C:\Users\Sriharshini\Downloads\nbfc_credit_risk_dataset.csv"
df = pd.read_csv(file_path)
```

```python
df.head()
```

```python
print("\nDataset Shape:", df.shape)
print("\nColumn Names:\n", df.columns)
print("\nData Types:\n")
print(df.dtypes)

df.describe(include="all")
```

```python
print("\nMissing Values:\n")
print(df.isnull().sum())
```

```python
df.describe().T
```

```python
num_cols = df.select_dtypes(include=["float64", "int64"]).columns

for col in num_cols[:6]:   # first 6 columns only (avoid too many plots)
    plt.figure()
    sns.histplot(df[col], kde=True, bins=30)
    plt.title(f"Distribution: {col}")
    plt.show()
```

```python
for col in num_cols[:6]:   # plot first 6 numeric columns
    plt.figure()
    sns.boxplot(x=df[col])
    plt.title(f"Outlier Check: {col}")
    plt.show()
```

```python
plt.figure(figsize=(14,10))

numeric_df = df.select_dtypes(include=["float64", "int64"])

corr = numeric_df.corr()

sns.heatmap(corr, cmap="coolwarm", annot=False)
plt.title("Correlation Heatmap (Numeric Columns Only)")
plt.show()
```

```python
sns.countplot(data=df, x="default")
plt.title("Default (0 = No, 1 = Yes)")
plt.show()

df["default"].value_counts(normalize=True)
```

```python
print("\nAverage DTI Ratio:", df["dti_ratio"].mean())
print("Average DSCR:", df["dscr"].mean())
print("Avg Past Delinquencies:", df["past_delinquencies"].mean())
print("Default Rate:", df["default"].mean())
```
